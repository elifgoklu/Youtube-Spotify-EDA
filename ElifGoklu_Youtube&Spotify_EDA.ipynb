### <font color='SlateBlue'> LET'S GET TO KNOW THE DATA: </font>


*This dataset contains statistical information about songs by world-renowned artists, including metrics for listens, views, and likes on YouTube and Spotify platforms.

**<font color='DarkBlue'>Track**: The name of the song as seen on the Spotify platform.
    
**<font color='DarkBlue'>Artist**: The name of the artist.
    
**<font color='DarkBlue'>Url_spotify**: The artist's URL.
    
**<font color='DarkBlue'>Album**: The album where the song is located.
    
**<font color='DarkBlue'>Album_Type**: Indicates whether the song was released as a single or as part of an album on Spotify.
    
**<font color='DarkBlue'>URL**: A Spotify link used to find the song via API.
    
**<font color='DarkBlue'>Danceability**: Defines a song's suitability for dancing based on a combination of musical elements such as tempo, rhythm stability, emphasis strength, and overall regularity. A value of 0.0 is the least danceable and 1.0 is the most danceable.
    
**<font color='DarkBlue'>Energy**: A measure between 0.0 and 1.0 representing the perceived intensity and activity of the song. Energetic tracks are usually fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low.

### üßº Data Cleaning

```python
# Removing the Unnecessary Unnamed Column
df= df.drop(["Unnamed: 0"], axis=1)
df.head(5)
```

```python
# Checking Null Values with Corresponding Percentages
df.isnull().sum()/len(df)
```

```python
# Filtering the Data to be Used
df = df[['Artist', 
    #'Url_spotify', 
    'Track', 'Album', 'Album_type', 
    #'Uri',
    'Danceability', 
    #'Energy', 'Key', 'Loudness', 
    #'Speechiness', 
    #'Acousticness', 'Instrumentalness', 'Liveness', 
    'Valence', 
    #'Tempo', 
    'Duration_ms', 
    #'Url_youtube', 
    'Title', 'Channel', 'Views', 'Likes', 'Comment Count', 
    #'Description', 'Licensed', 
    'Official Video', 'Stream']].copy()
```

```python
df.head(5)
```

```python
# Removing Null Values:
df.dropna(inplace=True)
```

```python
# Checking the Result
df.isnull().sum()
```

```python
# Checking for Duplicate Values
df.duplicated().sum()
```

## Data Preprocessing

```python
df.dtypes
```

```python
# Adjusting for Better Clarity in Visuals
df['Official Video'] = df['Official Video'].replace({True: 'Official', False: 'Not Official'})
```

```python
# Representing Integer Values as Integers for More Accurate Data Types
df[["Likes","Comment Count","Views","Stream"]]=df[["Likes","Comment Count","Views","Stream"]].astype("int64")
df.dtypes
```

```python
# Separating Numeric Data
numerical_df = df.select_dtypes(include=['float64',"int64"])
numerical_df
```

```python
# Calculating Song Durations in Minutes
df["Duration_min"] = (df['Duration_ms'] / 60000).round(2)
```

### üé§ üé∂ Top 10 Songs

```python
# Preparing the Parameters for the Top 10 Lists Based on Different Metrics
df_top10=df.groupby('Track').agg({"Likes":"sum",'Views':'sum',"Comment Count":"sum","Stream":"sum"})
df_top10['Rate'] = df_top10['Likes'] / df_top10['Views']
df_top10
```

```python
df_top10=df_top10.reset_index()
```

```python
# Top 10 Most Viewed Songs
df_views=df_top10.sort_values(by='Views',ascending=False).head(10)
# Top 10 Most Liked Songs
df_likes=df_top10.sort_values(by='Likes',ascending=False).head(10)
# op 10 Songs with the Highest Like-to-View Ratio
df_top10['Rate'] = df_top10['Likes'] / df_top10['Views']
df_rates=df_views.sort_values(by='Rate', ascending=False).head(10)
# Top 10 Most Commented Songs
df_comments=df_top10.sort_values(by='Comment Count', ascending=False).head(10)
# Top 10 Most Played Songs
df_stream=df_top10.sort_values(by='Stream', ascending=False).head(10)
df_stream
```

```python
# Most Played Songs
plt.figure(figsize=(12,7))
plt.subplot(2, 2, 1)
sns.barplot(x='Stream',y='Track',data=df_stream,color='Salmon')
plt.xticks(fontsize=8,rotation=0)
plt.yticks(fontsize=11,rotation=0)

# Most Viewed Songs
plt.subplot(2, 2, 2)
sns.barplot(x='Views',y='Track',data=df_views,color='Salmon')
plt.yticks(fontsize=11,rotation=0)
plt.xticks(fontsize=10,rotation=0)

# Most Played Songs by Number of Likes
plt.subplot(2, 2, 3)
sns.barplot(x='Likes',y='Track',data=df_likes,color='Salmon')
plt.yticks(fontsize=11,rotation=0)
plt.xticks(fontsize=10,rotation=0)

# Most Commented Songs
plt.subplot(2, 2, 4)
sns.barplot(x='Comment Count',y='Track',data=df_comments,color='Salmon')
plt.yticks(fontsize=11,rotation=0)
plt.xticks(fontsize=10,rotation=0)

plt.tight_layout()
plt.show()

      
```
